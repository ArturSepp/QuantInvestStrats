Factsheet for performances of multi assets

```python
# packages
import matplotlib.pyplot as plt
import yfinance as yf
import qis as qis
from qis import generate_multi_asset_factsheet, fetch_default_report_kwargs
```

```python
# define strategy universe data
benchmark = 'SPY'
tickers = [benchmark, 'QQQ', 'EEM', 'TLT', 'IEF', 'LQD', 'HYG', 'SHY', 'GLD']
time_period = qis.TimePeriod('31Dec2007', '16Oct2024')  # time period for reporting

prices = yf.download(tickers=tickers, start="2003-12-31", end=None, ignore_tz=True, auto_adjust=True)['Close'][tickers]
prices = prices.asfreq('B', method='ffill')  # make B frequency
prices = prices.dropna()
prices
```

```python
# run report
kwargs = fetch_default_report_kwargs(time_period=time_period)

fig = generate_multi_asset_factsheet(prices=prices,
                                         benchmark=benchmark,
                                         time_period=time_period,
                                         **kwargs)
```
